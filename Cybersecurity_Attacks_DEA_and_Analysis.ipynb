```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("00_cybersecurity_attacks.csv")

df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# print(df.isnull().sum())

df['Attack Type'].fillna('Unknown', inplace=True)
df['Action Taken'].fillna('Unknown', inplace=True)

severity_order = ['Low', 'Medium', 'High', 'Critical', 'Unknown']
df['Severity Level'] = pd.Categorical(df['Severity Level'], categories=severity_order, ordered=True)

attack_type_counts = df['Attack Type'].value_counts()
severity_counts = df['Severity Level'].value_counts()
action_taken_counts = df['Action Taken'].value_counts()

print("--- Attack Type Top 5 ---")
print(attack_type_counts.head(5))
print("\n--- Severity Level Distribution ---")
print(severity_counts)
print("\n--- Action Taken Distribution ---")
print(action_taken_counts)
```
